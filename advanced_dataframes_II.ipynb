```python
# Copy the users and roles DataFrames from the examples above.
# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles

```

```python
# What is the result of using a right join on the DataFrames?
# Perform an outer join specifying the left and right DataFrame keys.

users.merge(roles, right_on='id', left_on='role_id', how='inner', indicator=True)
```

```python
# What is the result of using an outer join on the DataFrames?
users.merge(roles, right_on='id', left_on='role_id', how='outer', indicator=True)
```

```python
# What happens if you drop the foreign keys from the DataFrames and try to merge them?
users.merge(roles, right_on='id', left_on='role_id', how='outer', indicator=True).drop(columns = ['id_y', 'role_id'])

# (users.merge(roles, 
#             left_on='role_id', 
#             right_on='id', 
#             how='outer')
#     .drop(columns='role_id')
#     .rename(columns={'id_x': 'id', 
#                      'name_x': 'employee',
#                      'id_y': 'role_id',
#                      'name_y': 'role'}
#             )
# )
```

```python
# Create a column named mileage_difference like you did in the DataFrames exercises; this column should 
# contain the difference between highway and city mileage for each car

mpg['mileage_difference']= mpg['hwy']- mpg['cty']
mpg
```

```python
# Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the city and highway mileage.
col =mpg.loc[: , "cty":"hwy"]
mpg['mileage_mean'] = col.mean(axis=1)
mpg
```
