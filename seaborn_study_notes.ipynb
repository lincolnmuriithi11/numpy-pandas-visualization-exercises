# Categorical plots
distributions using categorical data in reference to numerical data



Bar plots

```python
sns.barplot(x = 'sex',y= 'total_bill', data=tips_df, estimator =np.std)
# on estimator you can use different statistical items to come up with what you need: cov,std,median
```

# count plot

```python
sns.countplot(x='sex', data=tips_df)
```

# boxplot
will show you quartiles of the data

```python
sns.boxplot(x="day",y= "total_bill", data = tips_df, hue="sex")
plt.legend(loc = 0)#used to move the legend
# hue helps create additional categorical seperation
# bar across the middle is the median 
# end of box extends to one standard deviation from the median
# whiskers extends to all the data asides from the ones in the box and what are considered as outliers
```

# violin plot
similar use to a box plot

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue= 'sex',split =True)
```

# strip plots
draws a scater plot representing different types of datapoints where one of them is categorical

```python
plt.figure(figsize=(8,5))
sns.stripplot(x= 'day',y='total_bill', data =tips_df, jitter=True, hue = "sex", dodge= True
             )
#jitter spreads the data points out while dodge will seperate the sex category 
```

# Swarm Plot

```python
# sns.violinplot(x='day', y = 'total_bill', data = tips_df)
sns.swarmplot(x='day', y = 'total_bill', data = tips_df)
```
