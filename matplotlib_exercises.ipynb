```python
import matplotlib.pyplot as plt
import math
# generate data
x = range(-100, 100)
y = [(x**2 - x + 2) for x in x]


plt.plot(x,y)

# add title
plt.title('$ y = x^2 - x +2$')

# add labels
plt.xlabel('$x$')
plt.ylabel('$y$')

# add annotation
plt.annotate('Origin', xy = (0,0), xytext = (0, 4000), arrowprops ={})
%config InlineBackend.figure_format = 'png'

# save the figure
plt.savefig('figure1.png')


    
```

```python
# generate data
x1 = range(0,100)
y1 = [math.sqrt(n) for n in x1]

# plot the data
plt.plot(x1,y1)

# customize the chart
plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)

plt.grid(True, ls = '--');
```

```python
import matplotlib.pyplot as plt
import math

x2 = range(0,20)
y2 = [math.pow(n,3) for n in x2]

plt.plot(x2,y2)

plt.title('$y= x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--');
```

```python
x3 = [x * 0.1 for x in range(0,150)]
y3 = [2 ** n for n in x3]

plt.plot(x3,y3)

plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.grid(True, ls = '--')
```

```python
# specify range of x values
x4 = [x * 0.1 for x in range(0,150)]
y4 = [1/(n+1) for n in x4]


# plot the data
plt.plot(x4,y4)


# Customize the plot
# Add title
plt.title('$ y = 1/(x+1)$')

# Add x and y labels
plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')


plt.grid(True, ls = '--')

plt.show()
```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.


plt.figure(figsize = (10,6))

#subplot 1
plt.subplot(221)

plt.plot(x1,y1)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,10)
plt.ylim(0, 10)
plt.grid(True, ls = '--')

#subplot 2
plt.subplot(222)

plt.plot(x2,y2)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--')


#subplot 3
plt.subplot(223)

plt.plot(x3,y3)
plt.grid(True, ls = '--')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.title('$y = 2^x$')

#subplot 4
plt.subplot(224)

plt.plot(x4,y4)
plt.title('$ y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$1/x$')
plt.grid(True, ls = '--')
plt.xlim(0,15)
plt.ylim(0)
plt.annotate("some text", xy = (2,0.1), xytext = (2, 0.6), arrowprops = {});

plt.suptitle('Subplots Answer')
plt.tight_layout()
```

```python
#Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

plt.figure(figsize = (8,6))

# plot the data
plt.plot(x1, y1, c='#FF0000', label='$\sqrt{x}$', alpha=1.0)
plt.plot(x2, y2, c='#0000FF', label='$x^{3}$', alpha=.8)
plt.plot(x3, y3, c= '#008000', label='$2^{x}$', alpha=.8)
plt.plot(x4, y4, c= '#800080', label='$1/(x+1)$', alpha=.8)

# adjust xlim and ylim
plt.ylim(0, 10)
plt.xlim(0, 10)

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.title('Combined Functions')
plt.legend()

plt.show()
```
