Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
from env import get_db_url
url = get_db_url("chipotle")
chip= pd.read_sql('SELECT * FROM orders', url )
```

```python
chip.info()
```

```python
chip["item_price"]= chip["item_price"].replace('$', '').astype(float)

chip.head()
```

```python
popular_items = chip.groupby('item_name').quantity.sum().nlargest(4)
popular_items
```

```python
revenue = chip[chip.item_name.isin(popular_items.index)].groupby('item_name').item_price.sum().reset_index()
revenue
```

```python
plt.figure(figsize = (8, 5))

sns.histplot(data = revenue, y= 'item_name', x = 'item_price')
```
