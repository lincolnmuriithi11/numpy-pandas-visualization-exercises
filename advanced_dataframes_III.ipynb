```python

 
sql = '''
SELECT 
    *
FROM
    employees e
        JOIN
    titles t ON t.emp_no = e.emp_no
        JOIN
    dept_emp de ON de.emp_no = t.emp_no
        JOIN
    departments d ON d.dept_no = de.dept_no
''' #clean up your code on sql before attaching it to pandas, additionally add filters to cater for current employeesif neccesary
dpt_info = pd.read_sql(sql, url)
dpt_info
```

```python

pd.crosstab(dpt_info.title, dpt_info.dept_name)
```
